```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# Dataset
data = {
    "Rainfall_mm":[400,600,300,700,500],
    "Fertilizer_kg":[50,75,40,90,60],
    "Crop_Yield_Tons":[2.5,3.8,1.8,4.5,3.1]
}
df = pd.DataFrame(data)
# 1. Pearson Correlation
corr = df["Rainfall_mm"].corr(df["Crop_Yield_Tons"])
print("Pearson Correlation:", corr)
# 2. Linear Regression Model
X = df[["Rainfall_mm"]]
y = df["Crop_Yield_Tons"]
model = LinearRegression()
model.fit(X, y)
# 3. Prediction
pred = model.predict([[550]])
print("Predicted Crop Yield:", pred[0])
# 4. R2 Score
y_pred = model.predict(X)
print("R2 Score:", r2_score(y, y_pred))
# 5. Scatter Plot with Regression Line
plt.scatter(X, y, color="blue")
plt.plot(X, y_pred, color="red")
plt.xlabel("Rainfall")
plt.ylabel("Crop Yield")
plt.title("Simple Linear Regression")
plt.show()
# 6. Residual Plot
residual = y - y_pred
plt.scatter(y_pred, residual)
plt.axhline(y=0, color="red")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
